```python
from annoy import AnnoyIndex
import random

def build_annoy_index(data, num_trees=10):

    # Determine the dimensionality of the data
    dimension = len(data[0])

    # Create an Annoy index
    index = AnnoyIndex(dimension, 'angular')

    # Add items to the index
    for i, vector in enumerate(data):
        index.add_item(i, vector)

    # Build the index
    index.build(num_trees)

    return index

def find_similar_items(index, vector, num_neighbors=5):

    return index.get_nns_by_vector(vector, num_neighbors)


#Example
# Generate some random data
num_items = 1000
dimensions = 40
data = [[random.gauss(0, 1) for _ in range(dimensions)] for _ in range(num_items)]

# Build the Annoy index
annoy_index = build_annoy_index(data)

# Define a query vector (e.g., the first item in the data)
query_vector = data[0]

# Find similar items
similar_items = find_similar_items(annoy_index, query_vector)

print("Indices of similar items:", similar_items)
print("Similar items vectors:")
for idx in similar_items:
    print(data[idx])
```
